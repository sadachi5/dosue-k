```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import csv
import function as func
```

```python
for i in range(263, 264, 1):
    word = list(str(i))
    word.insert(2, ".")
    start_freq = float("".join(word))
    
    path = "../result_data/sensitivity_result/start_{}GHz.csv".format(start_freq)
    sen = func.csv_to_array(path)
    
    fig, axes = plt.subplots(3,3,figsize=(16,10))
    
    fmt_op = "o"
    size_op = 3

    ax = axes[0][0]
    ax.errorbar(sen["freq"], sen["Tplate"], yerr = sen["Tplate_err"], fmt=fmt_op, ms=size_op)
    ax.set_title("")
    ax.set_xlabel("Frequency [GHz]")
    ax.set_ylabel("Tplate [K]")
    ax.grid()
    
    ax = axes[0][1]
    ax.errorbar(sen["freq"], sen["Tsys"], yerr = sen["Tsys_err"], fmt=fmt_op, ms=size_op)
    ax.set_title("")
    ax.set_xlabel("Frequency [GHz]")
    ax.set_ylabel("Tsys [K]")
    ax.grid()
    
    ax = axes[0][2]
    ax.errorbar(sen["freq"], sen["NEP"], yerr = sen["NEP_err"], fmt=fmt_op, ms=size_op)
    ax.set_title("")
    ax.set_xlabel("Frequency [GHz]")
    ax.set_ylabel("NEP [W/âˆšHz]")
    ax.grid()

    fig.tight_layout()
    plt.savefig("../result_figure/sensitivity_fugure/start_{}GHz.pdf".format(str(start_freq)))
```
