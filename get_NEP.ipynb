```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import csv
import function as func
```

```python
# Constants
v_c = 220.e+3 # [m/sec] speed of solar system
v_E = v_c # [m/sec] speed of earth
c = 299792458. # [m/sec] speed of light from wikipedia
k_B = 1.380649e-23 # [J/K] boltzmann constant
rbw = 3.e+2 # [Hz]
binwidth = 5.e+3 # [Hz]
TLN2 = 77 # [K]
```

```python
for i in range(180, 264, 1):
    word = list(str(i))
    word.insert(2, ".")
    start_freq = float("".join(word))
    print(start_freq)
    
    column = ["freq", "Tsys", "Tplate", "NEP", "Tsys_err", "Tplate_err", "NEP_err"]
    with open("../result_data/sensitivity_result/start_{}GHz.csv".format(start_freq), "w") as f:
        writer = csv.writer(f)
        writer.writerow(column)

    #start_freq = 26.3
    path = "../result_data/yfactor_result/start_{}GHz.csv".format(str(start_freq))
    yfactor = func.csv_to_array(path)

    initial = int(start_freq * 1.e+6 - 250)
    final = int(initial + 1.e+5)

    for i in range(initial, final, 2000):
        word = list(str(i))
        word.insert(2, ".")
        start = "".join(word)

        Gain = 0
        Trx = 0
        Gain_err = 0
        Trx_err = 0
        for _f, freq_0 in enumerate(yfactor["freq"]):
            if float(freq_0) * 1e+6 == float(start) * 1e+6:
                Gain = yfactor["Gain"][_f]
                Trx = yfactor["Trx"][_f]
                Gain_err = yfactor["Gain_err"][_f]
                Trx_err = yfactor["Trx_err"][_f]
                break

        signal = func.csv_to_array("../result_data/signal_power/start_{}GHz.csv".format(start))
        freq = signal["freq"][0]/1e+9
        Tsys = np.mean(signal["W_all"]/k_B/rbw + Trx)
        Tplate = Tsys - Trx
        NEP = np.power(2 * rbw, 0.5) * k_B * Tsys
        Tsys_err = np.sqrt(np.power(np.std(signal["W_all"])/len(signal["W_all"])**0.5/k_B/rbw, 2.) + np.power(Trx_err, 2.))
        Tplate_err = (Tsys_err ** 2. + Trx_err ** 2.) ** 0.5
        NEP_err = np.power(2 * rbw, 0.5) * k_B * Tsys_err
        
        with open("../result_data/sensitivity_result/start_{}GHz.csv".format(start_freq), "a") as f:
            writer = csv.writer(f)
            writer.writerow([freq, Tsys, Tplate, NEP, Tsys_err, Tplate_err, NEP_err])
```
